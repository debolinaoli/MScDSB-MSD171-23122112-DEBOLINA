```python
class PetStore:
    def __init__(self):
        self.pet_inventory = {}

    def store_pet_details(self, pet_name, pet_details):
        self.pet_inventory[pet_name] = pet_details
        print(f"Pet '{pet_name}' details added successfully!")

    def search_pet(self, pet_name):
        if pet_name in self.pet_inventory:
            return self.pet_inventory[pet_name]
        else:
            return f"Pet '{pet_name}' not found in the inventory."

    def sell_pet(self, pet_name):
        if pet_name in self.pet_inventory:
            del self.pet_inventory[pet_name]
            print(f"Sold pet '{pet_name}' successfully!")
        else:
            print(f"Pet '{pet_name}' not found in the inventory.")

    def list_all_pets(self):
        if self.pet_inventory:
            print("List of available pets:")
            for pet_name, pet_details in self.pet_inventory.items():
                print(f"Name: {pet_name}, Details: {pet_details}")
        else:
            print("No pets available in the inventory.")

# Create a menu-driven program for the PetStore
if __name__ == "__main__":
    pet_store = PetStore()
    
    while True:
        print("\nPetStore Menu:")
        print("1. Admin - Add Pet Details")
        print("2. User - Search for a Pet")
        print("3. User - Buy a Pet")
        print("4. User - List All Pets")
        print("5. Exit")
        
        choice = input("Enter your choice (1/2/3/4/5): ")
        
        if choice == "1":
            pet_name = input("Enter the pet's name: ")
            pet_details = input("Enter the pet's details: ")
            pet_store.store_pet_details(pet_name, pet_details)
        
        elif choice == "2":
            pet_name = input("Enter the pet's name to search: ")
            result = pet_store.search_pet(pet_name)
            print(result)
        
        elif choice == "3":
            pet_name = input("Enter the pet's name to buy: ")
            pet_store.sell_pet(pet_name)
        
        elif choice == "4":
            pet_store.list_all_pets()
        
        elif choice == "5":
            print("Exiting the PetStore program. Goodbye!")
            break
        
        else:
            print("Invalid choice. Please enter a valid option.")
```
