```python
import random

# Creating lists of items, quantities, and prices
items = ["FERTILISERS", "BANANA", "ALMOND", "EGGS", "APPLES", "MILK", "ONION", "MANGO", "OILSEEDS", "LEMONS"]
quantities = [random.randint(1, 10) for _ in range(10)]
prices = [round(random.uniform(1, 10), 2) for _ in range(10)]

# Create and write to a CSV file
with open('23122112.txt', 'w') as txtfile:
    txtfile.write("Item,Quantity,UnitPrice,TotalCost\n")

    for _ in range(100):
        random_item_index = random.randint(0, 9)
        item = items[random_item_index]
        quantity = quantities[random_item_index]
        price_per_unit = prices[random_item_index]
        total_cost = round(quantity * price_per_unit, 2)
        txtfile.write(f"{item},{quantity},{price_per_unit},{total_cost}\n")

# Read the text file and print results
with open('23122112.txt', 'r') as txtfile:
    next(txtfile)  # skip the header row

    rows = [line.strip().split(',') for line in txtfile]

    print("Results for the first 5 rows:")
    for row in rows[:5]:
        item, quantity, total_cost = row[0], row[1], row[3]
        print(f"Item: {item}, Quantity: {quantity}, Total Cost: {total_cost}")

    print("\nResults for the last 10 rows:")
    for row in rows[-10:]:
        item, quantity, total_cost = row[0], row[1], row[3]
        print(f"Item: {item}, Quantity: {quantity}, Total Cost: {total_cost}")

    # Calculate and print total cost for each unique item
    item_totals = {}
    for row in rows:
        item, quantity, total_cost = row[0], int(row[1]), float(row[3])
        if item in item_totals:
            item_totals[item][0] += quantity
            item_totals[item][1] += total_cost
        else:
            item_totals[item] = [quantity, total_cost]

    print("\nTotal cost for each unique item:")
    for item, (total_quantity, total_cost) in item_totals.items():
        print(f"Item Name: {item}, Total Quantity: {total_quantity}, Total Cost: {total_cost}")

    # Calculate and print min and max prices for each item sold
    item_prices = {}
    for row in rows:
        item, price_per_unit = row[0], float(row[2])
        if item in item_prices:
            item_prices[item].append(price_per_unit)
        else:
            item_prices[item] = [price_per_unit]

    print("\nMinimum and Maximum Prices for Each Item Sold:")
    for item, price_list in item_prices.items():
        min_price = min(price_list)
        max_price = max(price_list)
        print(f"Item: {item}, Min Price: {min_price}, Max Price: {max_price}")
```

```python
import csv
with open("sales.csv", "w", newline="") as csvfile:
    fieldnames = ["Item", "Quantity", "Price", "Total Cost"]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    
    writer.writeheader()
    
    total_costs = {}  # To store total cost for each unique item
    
    for _ in range(100):
        random_item_index = random.randint(0, 9)
        quantity = quantities[random_item_index]
        price = prices[random_item_index]
        item = items[random_item_index]
        
        # Calculate total cost for the row
        total_cost = round(quantity * price, 2)
        
        # Write row to CSV
        writer.writerow({"Item": item, "Quantity": quantity, "Price": price, "Total Cost": total_cost})
        
        # Update total cost for the unique item
        if item in total_costs:
            total_costs[item] += total_cost
        else:
            total_costs[item] = total_cost

# Display the total cost for the first 5 rows and last 10 rows
with open("sales.csv", newline="") as csvfile:
    reader = csv.DictReader(csvfile)
    print("Total Cost for First 5 Rows:")
    for _ in range(5):
        row = next(reader)
        print(f"{row['Item']}: {row['Total Cost']}")
    
    print("\nTotal Cost for Last 10 Rows:")
    rows = list(reader)[-10:]
    for row in rows:
        print(f"{row['Item']}: {row['Total Cost']}")

# Calculate and display total cost for each unique item
print("\nTotal Cost for Each Unique Item:")
for item, total_cost in total_costs.items():
    print(f"{item}: {total_cost}")

# Calculate and display minimum & maximum price for each item sold
print("\nMinimum and Maximum Price for Each Item Sold:")
item_prices = {item: [] for item in items}
with open("sales.csv", newline="") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        item = row["Item"]
        price = float(row["Price"])
        item_prices[item].append(price)

for item, prices in item_prices.items():
    min_price = min(prices)
    max_price = max(prices)
    print(f"{item}: Min Price: {min_price}, Max Price: {max_price}")
```
