Create a menu driven program that will create new orders and update the inventory accordingly,export the final data to the data..

```python
orders = {}  # Store orders in a dictionary
def create_order():
    order_id = input("Enter order ID: ")
    customer_name = input("Enter customer name: ")
    order_items = input("Enter order items (comma-separated): ").split(',')
    orders[order_id] = {
        "Customer Name": customer_name,
        "Order Items": order_items
    }
    print(f"Order {order_id} created successfully!")
def list_orders():
    if not orders:
        print("No orders found.")
    else:
        for order_id, order_info in orders.items():
            print(f"Order ID: {order_id}")
            print(f"Customer Name: {order_info['Customer Name']}")
            print(f"Order Items: {', '.join(order_info['Order Items'])}")
            print()
while True:
    print("\nMenu:")
    print("1. Create New Order")
    print("2. List Orders")
    print("3. Exit")
    
    choice = input("Enter your choice (1/2/3): ")
    
    if choice == '1':
        create_order()
    elif choice == '2':
        list_orders()
    elif choice == '3':
        break
    else:
        print("Invalid choice. Please select a valid option (1/2/3).")
print("Thank you for using the order management system!")


```

```python
import csv

class SportMart:
    def __init__(self, store_name):
        self.store_name = store_name
        self.inventory = {}
        self.orders = []

    def load_data(self, file_name):
        try:
            with open(file_name, 'r') as file:
                reader = csv.reader(file)
                for row in reader:
                    if row[0] == 'Inventory':
                        item, quantity = row[1], int(row[2])
                        self.inventory[item] = quantity
                    elif row[0] == 'Order':
                        order_id, customer, item, quantity = row[1], row[2], row[3], int(row[4])
                        self.orders.append((order_id, customer, item, quantity))
        except FileNotFoundError:
            print(f"{file_name} not found. Initializing empty data.")

    def save_data(self, file_name):
        with open(file_name, 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(['Inventory'])
            for item, quantity in self.inventory.items():
                writer.writerow(['', item, quantity])
            writer.writerow(['Orders'])
            for order in self.orders:
                writer.writerow(['', *order])

    def display_inventory(self):
        print(f"Inventory at {self.store_name}:")
        for item, quantity in self.inventory.items():
            print(f"{item}: {quantity}")

    def display_orders(self):
        print(f"Order History at {self.store_name}:")
        for order in self.orders:
            order_id, customer, item, quantity = order
            print(f"Order ID: {order_id}, Customer: {customer}, Item: {item}, Quantity: {quantity}")

    def create_order(self, order_id, customer, item, quantity):
        if item in self.inventory and self.inventory[item] >= quantity:
            self.orders.append((order_id, customer, item, quantity))
            self.inventory[item] -= quantity
            print(f"Order for {quantity} {item}(s) placed successfully.")
        else:
            print(f"Item {item} not available in sufficient quantity.")

    def update_inventory(self, item, quantity):
        if item in self.inventory:
            self.inventory[item] += quantity
        else:
            self.inventory[item] = quantity

# Initialize the SportMart store and load data from a CSV file
trinity_store = SportMart("Trinity Store")
trinity_store.load_data("trinity_store_data.csv")

# Menu-driven program
while True:
    print("\nSportMart - Trinity Store")
    print("1. Display Inventory")
    print("2. Display Order History")
    print("3. Create New Order")
    print("4. Update Inventory")
    print("5. Save and Exit")
    choice = input("Enter your choice (1/2/3/4/5): ")

    if choice == '1':
        trinity_store.display_inventory()
    elif choice == '2':
        trinity_store.display_orders()
    elif choice == '3':
        order_id = input("Enter Order ID: ")
        customer = input("Enter Customer Name: ")
        item = input("Enter Item Name: ")
        quantity = int(input("Enter Quantity: "))
        trinity_store.create_order(order_id, customer, item, quantity)
    elif choice == '4':
        item = input("Enter Item Name: ")
        quantity = int(input("Enter Quantity to add to inventory: "))
        trinity_store.update_inventory(item, quantity)
    elif choice == '5':
        trinity_store.save_data("trinity_store_data.csv")
        print("Data saved. Exiting...")
        break
```
